```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,100)
plt.plot(x,x,label='linear')
plt.legend()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot([10,20,30,40],[10,40,90,160],color='lightblue',marker='o')
ax.scatter([10,20,30,40],[10,20,30,40],color='darkgreen',marker='^')
ax.set_xlim(10,40)
# ax.set_ylim()
```
